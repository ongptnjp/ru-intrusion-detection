```python
# data from ISCX-IDS
import xml.etree.ElementTree as et 
import pandas as pd

parseXML = et.parse("./dataset/ISCXIDS/12SatJun.xml")
# print("parseXML", parseXML)

xroot = parseXML.getroot()

df_cols = ["appName", "totalSourceBytes", "totalDestinationBytes", "totalDestinationPackets"]
#            "totalSourcePackets", "sourcePayloadAsBase64",
#            "sourcePayloadAsUTF", "destinationPayloadAsBase64", "destinationPayloadAsUTF",
#            "direction", "sourceTCPFlagsDescription", "destinationTCPFlagsDescription",
#            "source", "protocolName", "sourcePort", "destination", "destinationPort",
#            "startDateTime", "stopDateTime", "Tag"]

out_df = pd.DataFrame(columns = df_cols)

def getvalueofnode(node):
    """ return node text or None """
    return node.text if node is not None else None

for node in xroot:

    s_appName = node.find("appName")

    s_totalSourceBytes = node.find("totalSourceBytes")

    s_totalDestinationBytes = node.find("totalDestinationBytes")
    
    s_totalDestinationPackets = node.find("totalDestinationPackets")

    out_df = out_df.append(
        pd.Series([getvalueofnode(s_appName), getvalueofnode(s_totalSourceBytes),
                   getvalueofnode(s_totalDestinationBytes), getvalueofnode(s_totalDestinationPackets)],
                                     index = df_cols),
                           ignore_index = True)
    
print("complete")
```
